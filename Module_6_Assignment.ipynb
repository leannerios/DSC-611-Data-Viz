```python
#create an agent to find lat/lon based on address using geopy and put them in a dataframe

data=[]
def geo_function(city):

    address = city
    geolocator = Nominatim(user_agent="my_agent")
    location = geolocator.geocode(address)
    latitude = location.latitude
    longitude = location.longitude

    results = [address, latitude, longitude]

    data.append(results)
    df=pd.DataFrame(data, columns=('address', 'lat', 'lon'))

    return df

```

```python
geo_function('Boston,MA')
df=geo_function('Los Angeles,CA')
df=geo_function('New York, NY')
df
```

```python
df_sample = df.sample(n=100)
df_sample

#I should have added random.seed() so the sample and points of interest wouldn't change every time I ran this
```

```python
#create an empty map around your center
folium_map = folium.Map(location=[latitude, longitude], tiles="CartoDB positron", zoom_start=14,control_scale = True) #CHANGE tiles= to one of the other tilesets
folium_map
```

```python
# Define the colormap
colormap = cm.linear.RdYlGn_06.scale(0, 1)

# Create a legend element
legend_html = """
     <div style="position: fixed;
     bottom: 50px; left: 50px; width: 150px; height: 110px;
     border:2px solid grey; z-index:9999; font-size:14px;
     ">
     &nbsp; <b>Legend</b> <br>
     &nbsp; Cluster Size: <br>
     """

#Label the colors and range
legend_html += f"""
        &nbsp; <i class="fa fa-circle" style="color:{colormap(3/4)};"></i>
        &nbsp; 1 - 9<br>
        &nbsp; <i class="fa fa-circle" style="color:{colormap(1/3)};"></i>
        &nbsp; 10 - 99<br>
        &nbsp; <i class="fa fa-circle" style="color:{colormap(1/5)};"></i>
        &nbsp; 100+<br>

        """

legend_html += """
     </div>
     """


#Make another legend for the terrain map but with a fill color so its easier to see
legend_html2 = """
     <div style="position: fixed;
     bottom: 50px; left: 50px; width: 150px; height: 110px;
     border:2px solid grey; z-index:9999; font-size:14px;background-color:white;
     ">
     &nbsp; <b>Legend</b> <br>
     &nbsp; Cluster Size: <br>
     """

legend_html2 += f"""
        &nbsp; <i class="fa fa-circle" style="color:{colormap(3/4)};"></i>
        &nbsp; 1 - 9<br>
        &nbsp; <i class="fa fa-circle" style="color:{colormap(1/3)};"></i>
        &nbsp; 10 - 99<br>
        &nbsp; <i class="fa fa-circle" style="color:{colormap(1/5)};"></i>
        &nbsp; 100+<br>

        """

legend_html2 += """
     </div>
     """
```

```python
# These three lines should create FastMarkerClusters (useful for a large number of points)
# this code points to the sample_df
from folium.plugins import FastMarkerCluster
FastMarkerCluster(data=list(zip(df_sample['LAT'].values, df_sample['LON'].values))).add_to(folium_map)
folium.LayerControl().add_to(folium_map)
folium_map.get_root().html.add_child(folium.Element(legend_html))

for index, row in df_sample.iterrows():
    folium.CircleMarker(location=(row["LAT"],row["LON"]),popup=(row["NAME"]),color="#007849",fill=False).add_to(folium_map)
# view map
folium_map # zoom into NYC to see the POI
```

```python
#create an empty map around your center
folium_map_terrain = folium.Map(location=[latitude, longitude], tiles="OpenTopoMap", zoom_start=16,control_scale = True)
folium_map_terrain
```

```python
FastMarkerCluster(data=list(zip(df_sample['LAT'].values, df_sample['LON'].values))).add_to(folium_map_terrain)
folium.LayerControl().add_to(folium_map_terrain)
folium_map_terrain.get_root().html.add_child(folium.Element(legend_html2))

for index, row in df_sample.iterrows():
    folium.CircleMarker(location=(row["LAT"],row["LON"]),popup=(row["NAME"]),color="#007849",fill=False).add_to(folium_map_terrain)
# view map
folium_map_terrain # zoom into NYC to see the POI
```
